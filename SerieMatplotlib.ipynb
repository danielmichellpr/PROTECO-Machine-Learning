
1) Realiza una gráfica con x contra x, x al cuadrado y x cúbica.

```python
x = np.linspace(0,2,15)
y1 = x
y2 = x*x
y3 = x**3
plt.plot(x,y1,label='Lineal',markersize=5)
plt.plot(x,y2,label='Cuadrática',markersize=5)
plt.plot(x,y3,label='Cubica',markersize=5)
plt.title("Gráficas", size=17)
plt.xlabel('Eje x',size=14)
plt.ylabel('Eje y',size=14)
plt.legend()
plt.show()
```

2) Gráfica la función sen de manera que se vea parecido a la siguiente imagen(consideren la linea punteada y la cuadricula):

```python
x = np.arange(0.,10.,0.2)
seno = np.sin(x)
plt.plot(x,seno,'ob',label='Seno(x)',markersize=5)
plt.title("Función trigonométrica", size=17)
plt.xlabel('Eje x',size=14)
plt.ylabel('Eje y',size=14)
plt.legend()
plt.show()
```

```python
z=np.random.randint(1000,size= (1000,2))

df = pd.DataFrame(z)
df.columns = ['x1','x2']
# Creamos una función que aplique un condicional para obtener los 4 puntos de la imagen
def my_fun(x, x1, x2, Classification):
    #print (x)
    if (x[x1]> 500 and x[x2]>500):
        x[Classification]=1
    elif(x[x1]<=500 and x[x2]<=500):
        x[Classification]=1
    else:
        x[Classification]=0
    return x    

#Aplicacmos la función y a la agregamos a una nueva columna de clasificación
df_new =df.apply(lambda x: my_fun(x, 'x1', 'x2','Classification'), axis=1)
df_new

X1 = df_new.x1.values   
X2 = df_new.x2.values
X= df_new.iloc[:,:-1].values
Y = df_new.Classification.values 
plt.scatter(X1,X2, c=Y)
plt.xlabel('x1'), plt.ylabel('x2')
plt.title('Datos', size=15)
plt.show()
```
