# Clase3 Matplotlib y Pandas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=np.arange(0.,11.,1)
y_lineal = x

plt.plot(x,y_lineal,'--r',label='Lineal',markersize=5)   #Parametros de plot, "eje x", "eje y",
plt.title("Primera Gráfica Lineal", size=14)
plt.xlabel("Eje x", size=13)
plt.ylabel('Eje y', size=13)
plt.grid(True)        #Para tener una cuadricula, recibe un valor booleano
plt.legend()        #Se coloca el plt.legend() para que, para que se observe el label
plt.show()
```

```python
x=np.arange(0.,11.,1)
y_lineal = x

plt.plot(x,y_lineal,'or',markersize=10)   #Parametros de plot, "eje x", "eje y",
plt.title("Segunda Gráfica Lineal", size=14)
plt.xlabel("Eje x", size=13)
plt.ylabel('Eje y', size=13)
plt.grid(True)        #Para tener una cuadricula, recibe un valor booleano
plt.show()
```

```python
x=np.arange(0.,11.,1)
y_lineal = x
y_cuadrada = x**2
plt.plot(x,y_lineal,'+-r',label='Lineal',markersize=10)   #Parametros de plot, "eje x", "eje y",
plt.plot(x,y_cuadrada,'--g',label='Cuadrada',linewidth=3)
plt.title("Graficas", size=14)
plt.xlabel("Eje x", size=13)
plt.ylabel('Eje y', size=13)
plt.grid(True)        #Para tener una cuadricula, recibe un valor booleano
plt.legend()        #Se coloca el plt.legend() para que, para que se observe el label
plt.show()
```

```python
x=np.arange(0.,10.,0.2)
seno = np.sin(x)
coseno = np.cos(x)

plt.plot(x,seno,label='seno')
plt.plot(x,coseno,label='coseno')
plt.title('Funciones Trigonométricas')
plt.legend()
plt.grid(True,linewidth=1)
plt.show()
```

```python
monedas = {'UnPeso':8,'DosPesos':3,'CincoPesos':6,'DiezPesos':2}
nombreMonedas=list(monedas.keys())
numeroMonedas=list(monedas.values())
plt.bar(nombreMonedas,numeroMonedas,color='cyan',edgecolor='black',linewidth=5)
plt.title('Monedas (barras)')
plt.show()
```

```python
#plt.plot(nombreMonedas,numeroMonedas)      #Con plt.plot me une los puntos
plt.scatter(nombreMonedas,numeroMonedas, color='green',marker='o',s=50)
plt.title('Monedas (Dispesión)')
plt.show()
```

```python
plt.pie(numeroMonedas,labels=nombreMonedas,colors={'red','cyan','yellow','green'},autopct='%.2f%%',startangle=90) #El número de decimales que quiero que me muestre
plt.title('Monedas(pastel)')
plt.show()
```

### Gráfica de dispersión

```python


N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
area=(30*np.random.rand(N))**2

plt.scatter(x,y,s=area,c=colors,alpha=0.5)  #alpha nos cambia la transparencia de los puntos
plt.show()
```

### Gráficas juntas 

```python
#fig es el contenedor de la información, no es necesario ponerle fig, puede ser cualquier cosa para almacenar, nomralmente se usa fig
fig, (grafica1,grafica2) =plt.subplots(1,2,sharey=True, figsize=(10,4)) #sharey = Tus graficas comparten los mismos ejes
grafica1.bar(nombreMonedas,numeroMonedas)
grafica2.scatter(nombreMonedas,numeroMonedas)
plt.show()
```

### 3D

```python
from mpl_toolkits.mplot3d import Axes3D   #Llamamos la librería e importamos algo de ahí
```

```python
fig = plt.figure()    #Para almacenar ahí la información, prepara el escenario
ax = Axes3D(fig)       #Prepara los ejes 

N=100
x=np.random.rand(N)
y=np.random.rand(N)
z=np.random.rand(N)

ax.scatter(x,y,z, color='red')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')
ax.set_zlabel('Eje z')
plt.show()
```

```python
def z(x,y):
    return np.cos(np.sqrt(x**2+y**2))
```

```python
fig = plt.figure()    #Para almacenar ahí la información, el contenedor
ax = Axes3D(fig)   

x=np.linspace(-4,4,50)
y=np.linspace(-4,4,50)
X,Y = np.meshgrid(x,y) #Nos permite juntar la información, la variable x y la y, nos acondiciona para poder pasar a la función

ax.plot_surface(X,Y,z(X,Y), color = 'cyan')
plt.show()
```

# Pandas

## Series (Columnas) y DataFrame (Columnas y Filas)

### Series (Columnas)

```python
s = pd.Series()
print(s)
```

```python
#Convertimos numpy a pd.Series

saludo= np.array(['Hola','cómo','están'])   #Es un arreglo de numpy, pero yo quiero uno de pandas
s=pd.Series(saludo) #Trasformación de numpy a pandas

#Por defecto empieza en 0, pero podemos decirle desde donde quiero mis índices
s=pd.Series(saludo,index=[100,101,102])
print(s)
```

```python
#Convertimos diccionario a serie, los índices los pone como las llaves

diccionario = {'h':1,'o':2,'l':3,'a':4}
s=pd.Series(diccionario)
print(s['h'])
```

```python
#Combinar paqueterías, no es recomndable, pero se puede
x=np.array([1,2,3,4])
y=pd.Series([1,2,3,4])

plt.plot(x,y)
plt.show()
```
