```python
#Dataset para concatenar dos dataset


# Se crea un nuevo dataset
States = ['NY', 'NY', 'NY', 'NY', 'FL', 'FL', 'GA', 'GA', 'FL', 'FL'] #Le asignamos los estados 
data = [1.0, 2, 3, 4, 5, 6, 7, 8, 9, 10]                              #Los datos del df
idx = pd.date_range('1/1/2012', periods=10, freq='MS')                #Los índices ordenados por fechas
df1 = pd.DataFrame(data, index=idx, columns=['Revenue'])              #Creamos el df
df1['State'] = States                                                 #Agregamos otra columna llamada 'State' al df

#Se crea el segundo dataframe con similares instrucciones
data2 = [10.0, 10.0, 9, 9, 8, 8, 7, 7, 6, 6]
idx2 = pd.date_range('1/1/2013', periods=10, freq='MS')
df2 = pd.DataFrame(data2, index=idx2, columns=['Revenue'])
df2['State'] = States
```

```python
#Imprimimos por separado para observar la información 
df1
```

```python
#Imprimimos por separado
df2
```

```python
#Concatemamos los df
df = pd.concat([df1,df2])
df
```
